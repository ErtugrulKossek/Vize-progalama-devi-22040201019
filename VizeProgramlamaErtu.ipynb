```python
import random
import matplotlib.pyplot as plt
import math

def atis_hesapla():
    hedef_mesafe = 20000 + 200 * random.randint(-10, 10)
    hedef_genislik = 1000 + 100 * random.randint(-2, 2)
    hedef_sonu = hedef_mesafe + hedef_genislik
    hedef_yukseklik = 19
    ust_konum = [0, hedef_yukseklik]
    aci = math.radians(30) 
    min_hiz = 330
    max_hiz = 1800

    atis_sayisi = 0
    atis_hizi = (min_hiz + max_hiz) / 2
    atis_mesafeleri = []
    atis_yukseklikleri = []

    while True:
        atis_sayisi += 1
        t = (atis_hizi * math.sin(aci)) / 9.8
        atis_mesafe = atis_hizi * math.cos(aci) * t

        if atis_mesafe < hedef_mesafe:
            print(f"onune dustu")
            min_hiz = atis_hizi
            atis_hizi = (atis_hizi + max_hiz) / 2
        elif atis_mesafe > hedef_sonu:
            print(f"uzagına dustu")
            max_hiz = atis_hizi
            atis_hizi = (atis_hizi + min_hiz) / 2
        else:
            print(f"{atis_sayisi}. seferde vuruş gerçekleşmiştir hedefi vurmak için gereken hız: {atis_hizi:.2f}")
            break

        atis_mesafeleri.append(atis_mesafe)
        atis_yukseklikleri.append(hedef_yukseklik)

    plt.plot(atis_mesafeleri, atis_yukseklikleri, 'ro-')
    plt.xlabel("Mesafe (metre)")
    plt.ylabel("Yükseklik (metre)")
    plt.title("Başarılı Atışın Yükseklik-Mesafe Grafiği")

    plt.show()

    return atis_sayisi, atis_hizi, ust_konum

atis_hesapla()
```
